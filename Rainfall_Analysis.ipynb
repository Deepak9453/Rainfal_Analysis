```python
import pandas as pd

# 1. Import the dataset
df = pd.read_csv("rainfall1627650742214.csv")
```

```python
df
```

```python

# 2. Check for missing values and drop rows with any missing values
print("Missing values per column:\n", df.isnull().sum())
df = df.dropna()
```

```python
# 3. Find the district with the highest annual rainfall
max_district = df.loc[df['ANNUAL'].idxmax(), ['DISTRICT', 'STATE_UT_NAME', 'ANNUAL']]
print("District with highest annual rainfall:\n", max_district)
```

```python
# 4. Display the top 5 states with the highest annual rainfall
top_states = df.groupby('STATE_UT_NAME')['ANNUAL'].mean().sort_values(ascending=False).head(5)
print("Top 5 states with highest annual rainfall:\n", top_states)
```

```python

# # 5. Drop the seasonal columns

# # Drop the seasonal columns with exact names
# df = df.drop(columns=['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'])
```

```python
# 6. Display state-wise mean rainfall for all months using pivot table
pivot_mean = pd.pivot_table(df, index='STATE_UT_NAME', values=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'], aggfunc='mean')
print("State-wise mean rainfall (monthly):\n", pivot_mean)
```

```python

# 7. Display the count of districts in each state
district_count = df['STATE_UT_NAME'].value_counts()
print("Count of districts in each state:\n", district_count)
```

```python


# 8. For each state, display the district with highest rainfall in May
highest_may = df.loc[df.groupby('STATE_UT_NAME')['MAY'].idxmax(), ['STATE_UT_NAME','DISTRICT','MAY']]
print("District with highest May rainfall per state:\n", highest_may)
```
